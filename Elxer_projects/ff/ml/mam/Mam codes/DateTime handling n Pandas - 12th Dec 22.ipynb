```python
df = pd.read_csv(r'..\datafiles\bitcoindata.csv', parse_dates=True)
# Convert the column Date to DateTime Datatype
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.set_index('Date',  drop=True, inplace=True)
df
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] =df.index.day
df['WkDay'] = df.index.weekday
```

```python
df
```

```python
# difference in 1 unit period, This paramater you can change as required
df['Open'].diff()
```

```python
#moving average
df['RM3'] = df['Open'].rolling(3).mean()
df
```

```python
df[['WkDay', 'Open']].boxplot(by='WkDay', figsize=(15,5));
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Open'], marker='p', ms=14, mfc='r')
plt.plot(df['RM3'], marker='*', ms=16, mfc='m', ls = '--')
plt.plot(df['Open'].rolling(7).mean(), marker='o', ms=14, mfc='g', ls = '-.')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage import io
```

```python
image=io.imread(r"C:\Users\ADMIN\Downloads\2010.jpg",as_gray=True)
io.imshow(image);
```

```python
prob=io.imread(r"C:\Users\ADMIN\Downloads\2020.jpg",as_gray=True)
io.imshow(prob);
```

```python
print(image.shape)
print(prob.shape)
```

```python
image.min()
        
```

```python
image=image*255
```

```python
bins=np.arange(0,256,5)
plt.figure(figsize=(14,5))
plt.hist(image.ravel(),bins);
plt.xticks(bins,rotation=90);
plt.xlabel('bins');
plt.ylabel('Frequency');
```

```python
prob=prob*255
```

```python
bins=np.arange(0,256,5)
plt.figure(figsize=(14,5))
plt.hist(prob.ravel(),bins);
plt.xticks(bins,rotation=90);
plt.xlabel('bins');
plt.ylabel('Frequency');
```

```python
bins=np.arange(0,256,5)
image_freq,_=np.histogram(image,bins)
prob_freq,_=np.histogram(prob,bins)
```

```python
plt.plot(bins[1:],image_freq,color='m',marker='o');
plt.plot(bins[1:],prob_freq,color='r',marker='*');
plt.grid();

```

```python
h=np.array([10,15,1,2,3,5])
np.percentile(h,25)
np.percentile(h,50)
np.percentile(h,75)
```

```python
plt.boxplot(h,vert=False);
```

```python
#IQR = q3-q1
h=np.array([10,15,1,2,3,25])
IQR=np.percentile(h,75)-np.percentile(h,25)
upper_boundry=np.percentile(h,75)+1.5*IQR  #incase of upper limit
lower_boundry=np.percentile(h,25)-1.5*IQR  #incase of lower limit
print(upper_boundry)
print(lower_boundry)
plt.boxplot(h,vert=False);
```

```python
#so here our upper outlier is 31 so if we give above 31 then see what happen
h=np.array([10,15,1,2,3,35])
plt.boxplot(h,vert=False);
plt.show()
```

```python
#so here our lower outlier is -15 so if we give lower than -15 then see what happen
h=np.array([-18,15,1,2,3,25])
plt.boxplot(h,vert=False);
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.iloc[:,[2,3]].values
y=df.iloc[:,-1].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=123)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
ypred=knn_model.predict(X_test)
print(knn_model.score(X_test,y_test))
```

```python
from sklearn.metrics import plot_confusion_matrix,classification_report
```

```python
plot_confusion_matrix(knn_model,X_test,y_test);
```

```python
print(classification_report(y_test,ypred))
```

### checking for p-values  

```python
from scipy.stats import f_oneway
```

```python
setosa_mask=df['species']=='setosa'
versicolor_mask=df['species']=='versicolor'
virginica_mask=df['species']=='virginica'
_,p=f_oneway(df[setosa_mask]['sepal_length'],
         df[versicolor_mask]['sepal_length'],
         df[virginica_mask]['sepal_length'])

_,p
```

```python
if p<=0.05:
    print("columns are corelated /dependent(reject H0)")
else:
    print("column are not corelated/independent(h0 holds true)")
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X=df.iloc[:,[3,4,5,6]].values
y=df.iloc[:,0].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=123)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
model=DecisionTreeClassifier(criterion='gini',random_state=123)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))
cols=['mass','width','height','color_score']
classes=['apple','mandarin','orange','lemon']
tree.plot_tree(model,feature_names=cols,class_names=classes,filled=True);
```

### Gini Index

```python
1-(1/6)**2-(5/6)**2
```

```python
# calculating the entropy
-1*(1/6*(np.log2(1/6))+5/6*(np.log2(5/6)))
```

```python
model=DecisionTreeClassifier(criterion='entropy')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
plt.figure(figsize=(15,10))
cols=['mass','width','height','color_score']
classes=['apple','mandarin','orange','lemon']
tree.plot_tree(model,feature_names=cols,class_names=classes,filled=True);
```

### Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfmodel=RandomForestClassifier(n_estimators=5)
rfmodel.fit(X_train,y_train).score(X_test,y_test)
```

```python
a=rfmodel.estimators_
```

```python
for i in a:
    plt.figure(figsize=(20,15))
    cols=['mass','width','height','color_score']
    classes=['apple','mandarin','orange','lemon']
    tree.plot_tree(i,feature_names=cols,class_names=classes,filled=True);
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=dict()
grid['n_estimators']=[5,10,50,100]
grid['criterion']=['gini','entropy']
grid['max_depth']=[3,5]
grid
```

```python
rf = RandomForestClassifier()
turned_model = GridSearchCV(rf,grid)
turned_model.fit(X_train,y_train)
turned_model.best_params_
```

```python
cols=['mass','width','height','color_score']
classes=['apple','mandarin','orange','lemon']
for i in range(rfmodel.estimators_):
    fig, axes=plt.subplot(figsize=(12,12))
    tree.plot_tree(rfmodel.estimators_[i]
                   feature_names=cols,
                   class_names=classes
                   filled=True,
                   ax=axes);
    
```

```python
lab=df['review'].values
```

```python
n_review=[]
for i in lab:
    tha=(TextBlob(i)).sentiment[0]
    if tha>0.2:
        j='Pos'
        n_review.append(j)
    elif tha<-0.2:
        t='Neg'
        n_review.append(t)
    else:
        r='Neutral'
        n_review.append(r)
    
```

```python
print(n_review)
```

```python
df['N_Review']=n_review
```

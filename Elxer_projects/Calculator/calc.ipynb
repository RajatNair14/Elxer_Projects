```python
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
```

```python
plan_price=int(input("Enter the plan price: "))
start_date = input("Enter the Start date of the plan (DD-MM-YYYY format): ")  
months=int(input("Enter Numbers of Months: "))
start_date=datetime.strptime(start_date, "%d-%m-%Y")    
end_date = start_date + relativedelta(months=months)
remaining = end_date - start_date 
print("Validity Of Plan",remaining.days,"days")
per_day_price=plan_price/remaining.days
print("Per Day Price of Plan: Rs",per_day_price)

choice=input("Enter the choice 1 for Existing plan only speed upgrade and 2 for upgrade plan + remaining amount")

### Existing plan only speed upgrade
if choice=="1":
    new_plan_price=int(input("Enter the New Plan price: "))
    current_date = datetime.now()
    days_used=start_date-current_date
    print("Plan Used Days: ",abs(days_used.days))
    days_used_amount=abs(days_used.days*per_day_price)
    print("Used Amount: ",days_used_amount)
    remaining_amount=plan_price-days_used_amount
    print("Remaining Amount: ",remaining_amount)
    new_per_day_price=new_plan_price/remaining.days
    print("Per Day Price of New Plan: Rs",new_per_day_price)
    days_extended=remaining_amount/new_per_day_price
    print("days: ",days_extended)
    days_extended=days_extended.__floor__()
    end_plan_date=datetime.now()+timedelta(days=days_extended)
    print("New Plan will expire on:",end_plan_date.date())

### upgrade plan + remaining amount
if choice=="2":
    new_plan_price=int(input("Enter the New Plan price: "))
    months=int(input("Enter Numbers of Months: "))
    current_date = datetime.now()
    days_used=start_date-current_date
    print("Plan Used Days: ",abs(days_used.days))
    days_used_amount=abs(days_used.days*per_day_price)
    print("Used Amount: ",days_used_amount)
    remaining_amount=plan_price-days_used_amount
    print("Ramaining Amount: ",remaining_amount)
    new_per_day_price=new_plan_price/remaining.days
    print("Per Day Price of New Plan: Rs",new_per_day_price)
    days_extended=remaining_amount/new_per_day_price
    print("days: ",months,"+",days_extended)
```
